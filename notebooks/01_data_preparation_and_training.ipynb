```python
# Notebook: Pr√©paration des Donn√©es et Entra√Ænement du Mod√®le
# Projet: Assistant Intelligent Tunisien NLP
# ============================================

# CELLULE 1: Installation des biblioth√®ques n√©cessaires
# ======================================================

"""
Installation de toutes les d√©pendances requises pour le projet
"""

!pip install pandas numpy scikit-learn matplotlib seaborn
!pip install transformers torch torchvision torchaudio
!pip install sentence-transformers
!pip install datasets
!pip install langdetect
!pip install streamlit
!pip install plotly
!pip install python-Levenshtein

print("‚úÖ Toutes les biblioth√®ques sont install√©es avec succ√®s!")

# ============================================
# CELLULE 2: Importation des biblioth√®ques
# ============================================

"""
Import de toutes les biblioth√®ques n√©cessaires pour l'analyse et le traitement
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from collections import Counter
import re
from typing import List, Dict, Tuple

# Pour le NLP
from transformers import (
    AutoTokenizer, 
    AutoModel,
    pipeline
)
import torch
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Pour le preprocessing
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Pour la d√©tection de langue
from langdetect import detect, LangDetectException

warnings.filterwarnings('ignore')

# Configuration de l'affichage
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("‚úÖ Toutes les biblioth√®ques import√©es avec succ√®s!")

# ============================================
# CELLULE 3: Chargement et exploration initiale des donn√©es
# ============================================

"""
Chargement du dataset et exploration initiale
"""

# Chargement des donn√©es
df = pd.read_csv('../data/tunisian_assistant_data.csv')
print("=" * 80)
print("üìä ANALYSE EXPLORATOIRE DES DONN√âES")
print("=" * 80)

# Informations g√©n√©rales
print("\n1Ô∏è‚É£ INFORMATIONS G√âN√âRALES:")
print(f"   ‚Ä¢ Nombre total d'exemples: {len(df)}")
print(f"   ‚Ä¢ Nombre de colonnes: {len(df.columns)}")
print(f"   ‚Ä¢ Colonnes: {list(df.columns)}")

# Aper√ßu des premi√®res lignes
print("\n2Ô∏è‚É£ APER√áU DES DONN√âES:")
print(df.head(10))

# Types de donn√©es
print("\n3Ô∏è‚É£ TYPES DE DONN√âES:")
print(df.dtypes)

# Statistiques descriptives
print("\n4Ô∏è‚É£ STATISTIQUES DESCRIPTIVES:")
print(df.describe(include='all'))

# Valeurs manquantes
print("\n5Ô∏è‚É£ VALEURS MANQUANTES:")
missing_values = df.isnull().sum()
if missing_values.sum() == 0:
    print("   ‚úÖ Aucune valeur manquante d√©tect√©e!")
else:
    print(missing_values[missing_values > 0])

# Distribution des cat√©gories
print("\n6Ô∏è‚É£ DISTRIBUTION DES CAT√âGORIES:")
print(df['categorie'].value_counts())

# Distribution des langues
print("\n7Ô∏è‚É£ DISTRIBUTION DES LANGUES:")
print(df['langue'].value_counts())

# Distribution du dialecte tunisien
print("\n8Ô∏è‚É£ PR√âSENCE DU DIALECTE TUNISIEN:")
print(df['dialecte_tunisien'].value_counts())

# ============================================
# CELLULE 4: Visualisations
# ============================================

"""
Cr√©ation de visualisations pour mieux comprendre les donn√©es
"""

# Configuration de la figure
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Distribution des cat√©gories
df['categorie'].value_counts().plot(kind='bar', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution des Cat√©gories', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Cat√©gorie')
axes[0, 0].set_ylabel('Nombre d\'exemples')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Distribution des langues
df['langue'].value_counts().plot(kind='pie', ax=axes[0, 1], autopct='%1.1f%%')
axes[0, 1].set_title('Distribution des Langues', fontsize=14, fontweight='bold')
axes[0, 1].set_ylabel('')

# 3. Dialecte tunisien
df['dialecte_tunisien'].value_counts().plot(kind='bar', ax=axes[1, 0], color='coral')
axes[1, 0].set_title('Pr√©sence du Dialecte Tunisien', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Dialecte Tunisien')
axes[1, 0].set_ylabel('Nombre d\'exemples')

# 4. Longueur des questions
df['question_length'] = df['question'].str.len()
df['question_length'].hist(bins=30, ax=axes[1, 1], color='lightgreen', edgecolor='black')
axes[1, 1].set_title('Distribution de la Longueur des Questions', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Longueur (caract√®res)')
axes[1, 1].set_ylabel('Fr√©quence')

plt.tight_layout()
plt.savefig('data_exploration.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Visualisations cr√©√©es et sauvegard√©es!")

# ============================================
# CELLULE 5: Analyse de texte approfondie
# ============================================

"""
Analyse approfondie du contenu textuel
"""

print("\n" + "=" * 80)
print("üìù ANALYSE DU CONTENU TEXTUEL")
print("=" * 80)

# Statistiques de longueur
print("\n1Ô∏è‚É£ STATISTIQUES DE LONGUEUR DES TEXTES:")
df['reponse_length'] = df['reponse'].str.len()
df['question_words'] = df['question'].str.split().str.len()
df['reponse_words'] = df['reponse'].str.split().str.len()

print(f"\n   Questions:")
print(f"   ‚Ä¢ Longueur moyenne: {df['question_length'].mean():.2f} caract√®res")
print(f"   ‚Ä¢ Longueur min: {df['question_length'].min()} caract√®res")
print(f"   ‚Ä¢ Longueur max: {df['question_length'].max()} caract√®res")
print(f"   ‚Ä¢ Nombre moyen de mots: {df['question_words'].mean():.2f}")

print(f"\n   R√©ponses:")
print(f"   ‚Ä¢ Longueur moyenne: {df['reponse_length'].mean():.2f} caract√®res")
print(f"   ‚Ä¢ Longueur min: {df['reponse_length'].min()} caract√®res")
print(f"   ‚Ä¢ Longueur max: {df['reponse_length'].max()} caract√®res")
print(f"   ‚Ä¢ Nombre moyen de mots: {df['reponse_words'].mean():.2f}")

# Mots les plus fr√©quents dans les questions
print("\n2Ô∏è‚É£ MOTS LES PLUS FR√âQUENTS DANS LES QUESTIONS:")
all_words = ' '.join(df['question'].str.lower()).split()
word_freq = Counter(all_words)
print("\n   Top 15 mots:")
for word, freq in word_freq.most_common(15):
    print(f"   ‚Ä¢ {word}: {freq} fois")

# Distribution par cat√©gorie et langue
print("\n3Ô∏è‚É£ CROISEMENT CAT√âGORIE √ó LANGUE:")
cross_tab = pd.crosstab(df['categorie'], df['langue'])
print(cross_tab)

# ============================================
# CELLULE 6: Nettoyage et pr√©paration des donn√©es
# ============================================

"""
Nettoyage et pr√©paration des donn√©es pour l'entra√Ænement
"""

print("\n" + "=" * 80)
print("üßπ NETTOYAGE DES DONN√âES")
print("=" * 80)

# Fonction de nettoyage
def clean_text(text: str, preserve_arabic: bool = True) -> str:
    """
    Nettoie le texte en pr√©servant l'arabe et les caract√®res sp√©ciaux tunisiens
    
    Args:
        text: Texte √† nettoyer
        preserve_arabic: Pr√©server les caract√®res arabes
    
    Returns:
        Texte nettoy√©
    """
    if pd.isna(text):
        return ""
    
    # Conversion en string si n√©cessaire
    text = str(text)
    
    # Suppression des espaces multiples
    text = re.sub(r'\s+', ' ', text)
    
    # Suppression des espaces en d√©but et fin
    text = text.strip()
    
    return text

# Application du nettoyage
print("   ‚Ä¢ Nettoyage des questions...")
df['question_clean'] = df['question'].apply(clean_text)

print("   ‚Ä¢ Nettoyage des r√©ponses...")
df['reponse_clean'] = df['reponse'].apply(clean_text)

# V√©rification post-nettoyage
print(f"\n‚úÖ Nettoyage termin√©!")
print(f"   ‚Ä¢ Questions nettoy√©es: {len(df[df['question_clean'] != ''])}/{len(df)}")
print(f"   ‚Ä¢ R√©ponses nettoy√©es: {len(df[df['reponse_clean'] != ''])}/{len(df)}")

# Suppression des lignes vides
initial_count = len(df)
df = df[(df['question_clean'] != '') & (df['reponse_clean'] != '')]
final_count = len(df)

print(f"   ‚Ä¢ Lignes supprim√©es: {initial_count - final_count}")
print(f"   ‚Ä¢ Dataset final: {final_count} exemples")

# ============================================
# CELLULE 7: Pr√©paration du mod√®le - Sentence Transformers
# ============================================

"""
Chargement et pr√©paration du mod√®le de similarit√© s√©mantique
Nous utilisons un mod√®le multilingue pour supporter FR/EN/AR
"""

print("\n" + "=" * 80)
print("ü§ñ CHARGEMENT DU MOD√àLE NLP")
print("=" * 80)

# Chargement du mod√®le multilingue
model_name = 'sentence-transformers/paraphrase-multilingual-mpnet-base-v2'
print(f"\n   ‚Ä¢ Chargement du mod√®le: {model_name}")
print("   ‚Ä¢ Ce mod√®le supporte 50+ langues incluant FR, EN, AR")

embedder = SentenceTransformer(model_name)

print("\n‚úÖ Mod√®le charg√© avec succ√®s!")

# ============================================
# CELLULE 8: Cr√©ation des embeddings
# ============================================

"""
G√©n√©ration des embeddings pour toutes les questions
Les embeddings sont des repr√©sentations vectorielles des textes
"""

print("\n" + "=" * 80)
print("üî¢ G√âN√âRATION DES EMBEDDINGS")
print("=" * 80)

# G√©n√©ration des embeddings pour les questions
print("\n   ‚Ä¢ G√©n√©ration des embeddings pour les questions...")
question_embeddings = embedder.encode(
    df['question_clean'].tolist(),
    show_progress_bar=True,
    batch_size=32
)

print(f"\n‚úÖ Embeddings g√©n√©r√©s!")
print(f"   ‚Ä¢ Forme: {question_embeddings.shape}")
print(f"   ‚Ä¢ Dimension: {question_embeddings.shape[1]}")
print(f"   ‚Ä¢ Type: {type(question_embeddings)}")

# Sauvegarde des embeddings
print("\n   ‚Ä¢ Sauvegarde des embeddings...")
np.save('question_embeddings.npy', question_embeddings)
print("   ‚úÖ Embeddings sauvegard√©s dans 'question_embeddings.npy'")

# ============================================
# CELLULE 9: Cr√©ation de la classe Chatbot
# ============================================

"""
Impl√©mentation de la classe Chatbot pour les r√©ponses
"""

class TunisianAssistantChatbot:
    """
    Chatbot assistant intelligent tunisien bas√© sur la similarit√© s√©mantique
    Supporte le fran√ßais, l'anglais et l'arabe (incluant le dialecte tunisien)
    """
    
    def __init__(self, 
                 df: pd.DataFrame, 
                 embeddings: np.ndarray,
                 model_name: str = 'sentence-transformers/paraphrase-multilingual-mpnet-base-v2'):
        """
        Initialisation du chatbot
        
        Args:
            df: DataFrame contenant les questions et r√©ponses
            embeddings: Embeddings pr√©-calcul√©s des questions
            model_name: Nom du mod√®le Sentence Transformer
        """
        self.df = df
        self.embeddings = embeddings
        self.embedder = SentenceTransformer(model_name)
        
    def detect_language(self, text: str) -> str:
        """
        D√©tecte la langue du texte
        
        Args:
            text: Texte √† analyser
            
        Returns:
            Code de langue (fr, en, ar)
        """
        try:
            lang = detect(text)
            return lang
        except LangDetectException:
            return 'unknown'
    
    def find_best_match(self, query: str, top_k: int = 3) -> List[Dict]:
        """
        Trouve les meilleures correspondances pour une requ√™te
        
        Args:
            query: Question de l'utilisateur
            top_k: Nombre de r√©sultats √† retourner
            
        Returns:
            Liste de dictionnaires contenant les r√©sultats
        """
        # G√©n√©ration de l'embedding de la question
        query_embedding = self.embedder.encode([query])
        
        # Calcul de la similarit√© cosinus
        similarities = cosine_similarity(query_embedding, self.embeddings)[0]
        
        # Obtention des indices des top_k r√©sultats
        top_indices = np.argsort(similarities)[-top_k:][::-1]
        
        # Pr√©paration des r√©sultats
        results = []
        for idx in top_indices:
            results.append({
                'question': self.df.iloc[idx]['question'],
                'reponse': self.df.iloc[idx]['reponse'],
                'categorie': self.df.iloc[idx]['categorie'],
                'langue': self.df.iloc[idx]['langue'],
                'similarity': similarities[idx],
                'dialecte_tunisien': self.df.iloc[idx]['dialecte_tunisien']
            })
        
        return results
    
    def chat(self, query: str, threshold: float = 0.5) -> str:
        """
        R√©pond √† une question de l'utilisateur
        
        Args:
            query: Question de l'utilisateur
            threshold: Seuil de similarit√© minimal
            
        Returns:
            R√©ponse du chatbot
        """
        # D√©tection de la langue
        detected_lang = self.detect_language(query)
        
        # Recherche de la meilleure correspondance
        matches = self.find_best_match(query, top_k=1)
        
        if matches[0]['similarity'] >= threshold:
            response = matches[0]['reponse']
            category = matches[0]['categorie']
            confidence = matches[0]['similarity']
            
            return f"{response}\n\nüìå Cat√©gorie: {category} | üéØ Confiance: {confidence:.2%}"
        else:
            return self._generate_fallback_response(detected_lang)
    
    def _generate_fallback_response(self, lang: str) -> str:
        """
        G√©n√®re une r√©ponse par d√©faut selon la langue
        
        Args:
            lang: Code de langue
            
        Returns:
            Message par d√©faut
        """
        fallback_responses = {
            'fr': "D√©sol√©, je n'ai pas trouv√© de r√©ponse pr√©cise √† votre question. "
                  "Pourriez-vous reformuler ou poser une question diff√©rente ?",
            'en': "Sorry, I couldn't find a precise answer to your question. "
                  "Could you rephrase or ask a different question?",
            'ar': "ÿπÿ∞ÿ±ÿßŸãÿå ŸÑŸÖ ÿ£ÿ¨ÿØ ÿ•ÿ¨ÿßÿ®ÿ© ÿØŸÇŸäŸÇÿ© ŸÑÿ≥ÿ§ÿßŸÑŸÉ. "
                  "ŸáŸÑ ŸäŸÖŸÉŸÜŸÉ ÿ•ÿπÿßÿØÿ© ÿµŸäÿßÿ∫ÿ© ÿßŸÑÿ≥ÿ§ÿßŸÑ ÿ£Ÿà ÿ∑ÿ±ÿ≠ ÿ≥ÿ§ÿßŸÑ ŸÖÿÆÿ™ŸÑŸÅÿü"
        }
        
        return fallback_responses.get(lang, fallback_responses['fr'])

print("‚úÖ Classe Chatbot cr√©√©e avec succ√®s!")

# ============================================
# CELLULE 10: Test du Chatbot
# ============================================

"""
Tests du chatbot avec diff√©rentes questions
"""

print("\n" + "=" * 80)
print("üß™ TESTS DU CHATBOT")
print("=" * 80)

# Initialisation du chatbot
chatbot = TunisianAssistantChatbot(df, question_embeddings)

# Questions de test
test_questions = [
    "Comment r√©server un rendez-vous chez le m√©decin?",
    "Kifech nÿÆŸÑÿµ facture STEG?",
    "Where can I find apartments for rent?",
    "9adech le prix du m√©tro?",
    "What are the best universities in Tunisia?",
    "O√π acheter une voiture d'occasion?"
]

print("\nüîç Test avec diff√©rentes questions:\n")
for i, question in enumerate(test_questions, 1):
    print(f"\n{'='*60}")
    print(f"Question {i}: {question}")
    print(f"{'='*60}")
    response = chatbot.chat(question)
    print(f"\n{response}\n")

print("\n‚úÖ Tests termin√©s!")

# ============================================
# CELLULE 11: √âvaluation et m√©triques
# ============================================

"""
√âvaluation de la performance du syst√®me
"""

print("\n" + "=" * 80)
print("üìä √âVALUATION DU SYST√àME")
print("=" * 80)

# Split train/test
from sklearn.model_selection import train_test_split

# Cr√©ation d'un split 80/20
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['categorie'])

print(f"\n   ‚Ä¢ Ensemble d'entra√Ænement: {len(train_df)} exemples")
print(f"   ‚Ä¢ Ensemble de test: {len(test_df)} exemples")

# Recalcul des embeddings pour l'ensemble d'entra√Ænement
train_embeddings = embedder.encode(
    train_df['question_clean'].tolist(),
    show_progress_bar=True,
    batch_size=32
)

# Cr√©ation d'un nouveau chatbot sur les donn√©es d'entra√Ænement
train_chatbot = TunisianAssistantChatbot(train_df.reset_index(drop=True), train_embeddings)

# √âvaluation sur l'ensemble de test
correct_matches = 0
total_tests = len(test_df)

print(f"\n   ‚Ä¢ √âvaluation en cours sur {total_tests} exemples...")

for idx, row in test_df.iterrows():
    matches = train_chatbot.find_best_match(row['question_clean'], top_k=1)
    if matches[0]['categorie'] == row['categorie']:
        correct_matches += 1

accuracy = correct_matches / total_tests
print(f"\n‚úÖ R√©sultats:")
print(f"   ‚Ä¢ Pr√©cision (cat√©gorie correcte): {accuracy:.2%}")
print(f"   ‚Ä¢ Correspondances correctes: {correct_matches}/{total_tests}")

# ============================================
# CELLULE 12: Sauvegarde du mod√®le et des donn√©es
# ============================================

"""
Sauvegarde de tous les √©l√©ments n√©cessaires pour le d√©ploiement
"""

print("\n" + "=" * 80)
print("üíæ SAUVEGARDE DES DONN√âES ET DU MOD√àLE")
print("=" * 80)

# Sauvegarde du DataFrame nettoy√©
df.to_csv('tunisian_assistant_data_clean.csv', index=False)
print("   ‚úÖ DataFrame nettoy√© sauvegard√©: tunisian_assistant_data_clean.csv")

# Sauvegarde des embeddings
np.save('question_embeddings_final.npy', question_embeddings)
print("   ‚úÖ Embeddings sauvegard√©s: question_embeddings_final.npy")

# Sauvegarde de la configuration
import json

config = {
    'model_name': model_name,
    'embedding_dim': question_embeddings.shape[1],
    'num_examples': len(df),
    'categories': df['categorie'].unique().tolist(),
    'languages': df['langue'].unique().tolist(),
    'accuracy': accuracy
}

with open('model_config.json', 'w', encoding='utf-8') as f:
    json.dump(config, f, ensure_ascii=False, indent=2)

print("   ‚úÖ Configuration sauvegard√©e: model_config.json")

print("\n" + "=" * 80)
print("‚úÖ ENTRA√éNEMENT ET PR√âPARATION TERMIN√âS!")
print("=" * 80)
print(f"\nR√©capitulatif:")
print(f"   ‚Ä¢ Dataset: {len(df)} exemples")
print(f"   ‚Ä¢ Cat√©gories: {len(df['categorie'].unique())}")
print(f"   ‚Ä¢ Langues support√©es: FR, EN, AR (+ dialecte tunisien)")
print(f"   ‚Ä¢ Pr√©cision: {accuracy:.2%}")
print(f"   ‚Ä¢ Mod√®le: {model_name}")
print(f"\nüöÄ Le syst√®me est pr√™t pour le d√©ploiement!")

# ============================================
# CELLULE 13: Visualisation de la distribution des embeddings
# ============================================

"""
Visualisation de la distribution des embeddings avec t-SNE
"""

from sklearn.manifold import TSNE
import plotly.express as px

print("\n   ‚Ä¢ R√©duction de dimensionnalit√© avec t-SNE...")

# R√©duction √† 2 dimensions
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
embeddings_2d = tsne.fit_transform(question_embeddings[:200])  # Utiliser un sous-ensemble pour la vitesse

# Cr√©ation du DataFrame pour la visualisation
viz_df = pd.DataFrame({
    'x': embeddings_2d[:, 0],
    'y': embeddings_2d[:, 1],
    'categorie': df['categorie'][:200],
    'langue': df['langue'][:200],
    'question': df['question'][:200]
})

# Cr√©ation du graphique interactif
fig = px.scatter(
    viz_df, 
    x='x', 
    y='y', 
    color='categorie',
    hover_data=['question', 'langue'],
    title='Distribution des Questions dans l\'Espace S√©mantique',
    width=1000,
    height=700
)

fig.write_html('embeddings_visualization.html')
print("   ‚úÖ Visualisation sauvegard√©e: embeddings_visualization.html")

plt.figure(figsize=(12, 8))
for categorie in viz_df['categorie'].unique():
    mask = viz_df['categorie'] == categorie
    plt.scatter(viz_df[mask]['x'], viz_df[mask]['y'], label=categorie, alpha=0.6, s=100)

plt.title('Distribution des Questions dans l\'Espace S√©mantique (t-SNE)', fontsize=16, fontweight='bold')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('embeddings_tsne.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Toutes les visualisations sont cr√©√©es!")
print("\n" + "=" * 80)
print("üéâ NOTEBOOK TERMIN√â AVEC SUCC√àS!")
print("=" * 80)
```
